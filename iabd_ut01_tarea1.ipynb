 3. Define la clase Pais con un "constructor" que inicialice los atributos nombre,
 capital e idioma, que serán los datos a considerar en la clase. Dentro de ella, crea los métodos
 get_nombre, get_capital y get_idioma, que devolverán el nombre del país, su capital y el idioma
 hablado, respectivamente. A continuación, crea una instancia u objeto de la clase Pais,
 asignándole los datos que desees, y haz uso de los métodos implementados para mostrar un
 mensaje con la información almacenada.

```python
class pais:
  def __init__(self, nombre, capital, idioma):
    self.nombre = nombre
    self.capital = capital
    self.idioma = idioma
  def get_nombre(self):
    print('El nombre del país es: ' + self.nombre)
  def get_capital(self):
    print('La capital del país es: ' + self.capital)
  def get_idioma(self):
    print('El idioma del país es: ' + self.idioma)
mi_pais = pais('España', 'Madrid', 'Español')
pais.get_nombre(self = mi_pais )
pais.get_capital(self = mi_pais )
pais.get_idioma(self = mi_pais)
```
